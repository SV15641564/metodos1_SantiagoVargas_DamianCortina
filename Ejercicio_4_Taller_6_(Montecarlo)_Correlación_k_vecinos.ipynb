```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#VALORES

k=np.linspace(0,30, 30)
N=10000
x = np.random.uniform(0,1, N)
Sample = np.zeros(30)
```

```python
#SUMATORIA

for a in range(1, 30):
  suma=0
  #El límite superior del rango está dado por la expresión: s=N-len(k). 
  #Esto se hace para que los índices mayores a s no estén fuera de alcance en la operación x[i+a].
  for i in range (0, 9970):
    valor=(x[i]*x[i+a])
    suma+=valor
  if a ==1:
    Sample[0]=suma/(N)
  Sample[a]=suma/(N)
```

```python
print(Sample)
```

```python
plt.plot(k,Sample)
```

Nótese que los valores fluctuan alrededor del valor teórico C(k)=1/4.
